##  AI vs Machine Learning vs Deep Learning



### üîπ Artificial Intelligence (AI)
Artificial Intelligence is a broad field of computer science focused on creating machines that can mimic human intelligence. This includes tasks like problem-solving, decision-making, understanding language, and recognizing patterns. AI systems can be **rule-based** or **learning-based**, and they don't always require data to function.

**Example:** A rule-based chatbot that answers questions based on fixed rules.

---

### üîπ Machine Learning (ML)
Machine Learning is a subfield of AI that allows machines to learn from data and improve over time without being explicitly programmed. It uses algorithms to find patterns in data and make predictions or decisions.

**Example:** A spam filter that learns from past emails to detect spam messages.

---

### üîπ Deep Learning (DL)
Deep Learning is a subfield of Machine Learning that uses **artificial neural networks** with many layers (hence "deep"). It is especially powerful in handling complex data like images, audio, and text. Deep learning models often require large datasets and high computing power.

**Example:** A facial recognition system that identifies people from photos using neural networks.

#Rule Based Agent


## ü§ñ Rule-Based Tic-Tac-Toe Agent

We built a basic AI agent that scans each board position and checks for a win for 'x'.

### Logic:
- If any row, column, or diagonal contains 3 'x' ‚Üí Predict `'positive'`
- Else ‚Üí Predict `'negative'`

### Key Insight:
This is a classic example of a **deterministic rule-based agent**. It doesn‚Äôt learn from data, but directly checks winning conditions like a human would.

### Performance:
We compared the predictions with the actual labels and achieved accuracy using simple logic ‚Äî showing how effective symbolic AI can be in fixed environments like games.


```python
# Define column names
columns = ['top-left', 'top-middle', 'top-right',
           'middle-left', 'middle-middle', 'middle-right',
           'bottom-left', 'bottom-middle', 'bottom-right', 'class']

# Load dataset with proper names
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/tic-tac-toe/tic-tac-toe.data"
data = pd.read_csv(url, names=columns)

# Check the first few rows and column names
print(data.columns)
print(data.head())

# Function to check winning condition for 'x'
def rule_based_tictactoe(row):
    # Extract board positions
    board = [ row['top-left'], row['top-middle'], row['top-right'],
        row['middle-left'], row['middle-middle'], row['middle-right'],
        row['bottom-left'], row['bottom-middle'], row['bottom-right']]

    # Rows
    rows = [
        board[0:3],
        board[3:6],
        board[6:9]
    ]

    # Columns
    cols = [
        [board[0], board[3], board[6]],
        [board[1], board[4], board[7]],
        [board[2], board[5], board[8]]
    ]

    # Diagonals
    diags = [
        [board[0], board[4], board[8]],
        [board[2], board[4], board[6]]
    ]

    # All winning lines
    lines = rows + cols + diags

    # If any line has all 'x', then it's a win
    for line in lines:
        if all(cell == 'x' for cell in line):
            return 'positive'
    return 'negative'
data.shape
```

```python
data['predicted_class'] = data.apply(rule_based_tictactoe, axis=1)
```

```python
# Check how many predictions match the actual class
correct_predictions = (data['class'] == data['predicted_class']).sum()
print(f"Correct Predictions: {correct_predictions}")
# Calculate accuracy
accuracy = (data['class'] == data['predicted_class']).mean()
print(f"Rule-Based Accuracy: {accuracy:.2f}")
```

# Search Algorithms

## BFS vs DFS - Search Algorithms in AI
**DFS (Depth-First Search)** and **BFS (Breadth-First Search)** are search algorithms used to explore mazes, trees, or graphs in AI.

---

### üåê BFS ‚Äì Breadth-First Search
- Explores level-by-level (all neighbors first).
- Uses a **Queue (FIFO)**.
- ‚úÖ Finds **shortest path** in unweighted graphs.
- ‚ùóÔ∏è Uses more memory.

---

### üï≥Ô∏è DFS ‚Äì Depth-First Search
- Goes deep into one path before backtracking.
- Uses a **Stack (LIFO)** or recursion.
- ‚ùå Doesn't guarantee shortest path.
- ‚úÖ Uses less memory.

---

### üîÑ Quick Comparison

| Feature        | BFS            | DFS             |
|----------------|----------------|-----------------|
| Structure      | Queue (FIFO)   | Stack / Recursion |
| Path Type      | Shortest path  | Any path        |
| Memory         | Higher         | Lower           |
| Use Case       | Mazes, routes  | Puzzles, trees  |

---

### üí° Tip:
- Use **BFS** for shortest path problems.
- Use **DFS** for exploration or puzzles.


## Maze Solver


```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

maze = []

print("Enter the maze row by row (0=wall, 1=path, 2=goal, 3=start):" )
for i in range(rows):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    maze.append(row)

print(maze)  # Just print it to check for now
```

```python
start = None
goal = None

for i in range(rows):              # loop through each row
    for j in range(cols):         # loop through each column
        if maze[i][j] == 3:       # 3 means Start
            start = (i, j)
        elif maze[i][j] == 2:     # 2 means Goal
            goal = (i, j)
```

```python
print("Start position:", start)
print("Goal position:", goal)
```

### *BFS*

```python

def bfs(maze, start, goal):
    rows = len(maze)
    cols = len(maze[0])
    directions = [(-1,0), (1,0), (0,-1), (0,1)]  # up, down, left, right

    visited = set()
    queue = deque()
    queue.append((start, [start]))  # (current_position, path_so_far)

    while queue:
        current, path = queue.popleft()

        if current == goal:
            return path  # FOUND GOAL

        x, y = current
        for dx, dy in directions:
            nx, ny = x + dx, y + dy


            if 0 <= nx < rows and 0 <= ny < cols:        # inside maze
                if maze[nx][ny] != 0 and (nx, ny) not in visited:  # not wall, not visited
                    visited.add((nx, ny))                 # mark visited
                    queue.append(((nx, ny), path + [(nx, ny)]))  # add new path

    return None  # No path found
```

```python
path = bfs(maze, start, goal)
print("Path:", path)
```

```python
def draw_maze_with_path(maze, path):
    visual = [[' ' for _ in row] for row in maze]

    for i in range(len(maze)):
        for j in range(len(maze[0])):
            if maze[i][j] == 0:
                visual[i][j] = '#'
            elif maze[i][j] == 3:
                visual[i][j] = 'S'
            elif maze[i][j] == 2:
                visual[i][j] = 'G'
            else:
                visual[i][j] = ' '  # will be replaced if it's in path

    #  Now mark the path (excluding start and goal)
    for i in range(1, len(path)-1):  # skip start and goal
        x, y = path[i]
        px, py = path[i-1]

        if x == px:
            visual[x][y] = '_'  # moved horizontally
        else:
            visual[x][y] = '|'  # moved vertically

    #  Finally, print the visual grid
    print("\nMaze with Path:")
    for row in visual:
        print(" ".join(row))
```

```python
if path:
    draw_maze_with_path(maze, path)
else:
    print("No path found!")
```

### DFS


```python
def dfs(maze, current, goal, visited, path, rows, cols):
    x, y = current

    # Out of bounds or hitting a wall or already visited
    if (x < 0 or x >= rows or y < 0 or y >= cols or
        maze[x][y] == 0 or (x, y) in visited):
        return False

    # Mark current cell
    visited.add((x, y))
    path.append((x, y))

    # Goal check
    if current == goal:
        return True

    # Explore all directions
    directions = [(-1,0), (1,0), (0,-1), (0,1)]
    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if dfs(maze, (nx, ny), goal, visited, path, rows, cols):
            return True

    # Dead end: Show attempted path
    draw_maze_with_path(maze, path.copy())
    path.pop()
    return False
```

```python
visited = set()
path = []

if dfs(maze, start, goal, visited, path, rows, cols):
    print("\nFinal Path (DFS found the goal):")
    draw_maze_with_path(maze, path)
else:
    print("\nNo path found using DFS.")
```
